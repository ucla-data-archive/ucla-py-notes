```python
# plotting multiple files (loops + matplotlib + numpy)

import glob
import numpy
import matplotlib.pyplot
filenames = sorted(glob.glob('data/inflammation-*.csv'))
filenames = filenames[0:3]
print(filenames)

for i in filenames:
    print(i)
    my_data = numpy.loadtxt(fname=i, delimiter=',')
    
    my_figure = matplotlib.pyplot.figure(figsize=(10.0, 3.0))
    my_figure.suptitle(i)
    
    mean_subplot = my_figure.add_subplot(1, 3, 1)
    mean_subplot.plot(numpy.mean(my_data, axis=0))
    mean_subplot.set_ylabel("average")
    mean_subplot.set_title("Mean")
    mean_subplot.axhline(8, 0, 1)
    
    max_subplot = my_figure.add_subplot(1, 3, 2)
    max_subplot.plot(numpy.max(my_data, axis=0))
    max_subplot.set_ylabel("maximum")
    max_subplot.set_title("Max")
    max_subplot.axvline(20, 0.25, 0.75)
    
    min_subplot = my_figure.add_subplot(1, 3, 3)
    min_subplot.plot(numpy.min(my_data, axis=0))
    min_subplot.set_ylabel("minimum")
    min_subplot.set_title("Min")
    
    my_figure.tight_layout()
    
    matplotlib.pyplot.show()
```
